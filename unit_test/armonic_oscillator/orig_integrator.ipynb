```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
fe_df = pd.read_csv('fe_solution.csv')

fe_time = fe_df['time']
fe_position = fe_df['position']
fe_velocity = fe_df['velocity']
fe_exact = fe_df['exact_solution']
fe_error = fe_df['abs_error']

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(fe_time,fe_position,label='Forward Euler')
plt.plot(fe_time,fe_exact,label='Exact')
plt.title('Position over Time')
plt.xlabel('Time')
plt.ylabel('Position')
plt.legend()
plt.grid(True)
plt.show()
```
